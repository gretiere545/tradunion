```
df_lim = df.copy()
```

```
df_lim.dropna(subset=['Name'], inplace=True)
```

```
print(df_lim['Name'].iloc[10:20])
```

```
def get_summary(x):
  try:
    t = wikipedia.summary(x)
    if 'langue' in t:
      return t
    else:
      return ''
  except:
    return ''
```

```
dep = 0
fin = 0
inc = 10
max = df_lim['Name'].count()
print (str(max))
```

```
fin = fin + inc -1
df_lim['summary'] = np.nan
while max >= fin:
  print (str(dep) + '-' + str(fin))
  df_lim['summary'].iloc[dep:fin] = df_lim.iloc[dep:fin].apply(lambda x: get_summary(x['Name']), axis=1)
  dep = dep + inc
  fin = fin + inc

reste = fin-max
if reste > 0:
  print (str(dep) + '-' + str(fin-reste-1))
  df_lim['summary'].iloc[dep:fin-reste-1] = df_lim.iloc[dep:fin-reste-1].apply(lambda x: get_summary(x['Name']), axis=1)
  
```
